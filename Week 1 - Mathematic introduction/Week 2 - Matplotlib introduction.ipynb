# Matplotlib introduction

Matplotlib is a library for making 2D plots of arrays in Python. Although it has its origins in emulating the MATLAB® graphics commands, it is independent of MATLAB, and can be used in a Pythonic, object oriented way. Although Matplotlib is written primarily in pure Python, it makes heavy use of NumPy and other extension code to provide good performance even for large arrays. 

The Matplotlib code is conceptually divided into three parts: 

1)

the pylab interface is the set of functions provided by matplotlib.pylab which allow the user to create plots with code quite similar to MATLAB figure generating code. 

2)

The Matplotlib frontend or Matplotlib API is the set of classes that do the heavy lifting, creating and managing figures, text, lines, plots and so on (Artist tutorial). This is an abstract interface that knows nothing about output. 

3)

The backends are device-dependent drawing devices, aka renderers, that transform the frontend representation to hardcopy or a display device. Example backends: PS creates PostScript® hardcopy, SVG creates Scalable Vector Graphics hardcopy, Agg creates PNG output using the high quality Anti-Grain Geometry library that ships with Matplotlib, GTK embeds Matplotlib in a Gtk+ application, GTKAgg uses the Anti-Grain renderer to create a figure and embed it in a Gtk+ application, and so on for PDF, WxWidgets, Tkinter, etc.

Typically the easiest way to start with plots, is to use the ```pyplot``` collection of commands. This can be loaded with

````python
import matplotlib.pyplot as plt
````

One can plot something using ```plt.plot()``` very quickly

````python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
````

At the end of every set of commands, sometime you need to have ```plt.show()``` to actually visualise the plot itself. Of course we gave only $y$ values so ```matplotlib``` generated $x$ values for us. When not given, ```matplotlib``` will start generating the $x$ values starting from $0$. We can give also $x$ values of course

````python
plt.plot([1,3,7,8],[1, 2, 3, 4])
````

## Formatting the plot - the basics

After the $x$ and $y$ one can give ```matplotlib``` a string indicating the color of the elements of the plot, in particular of points. You can concatenate a **color string** with a **line style** string. For example to plot red points you would use ```'ro'```. ```r``` for red, and ```o``` for points. For a blue line for example you would use ```b-```.

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
````

To change the limits of the axis, you can use the command ```plt.axis()```. This functions wants a list as argument in this form

```[xmin, xmax, ymin, ymax]```

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
````

## Matplotlib and numpy

Now ```matplotlib``` shines because it works beatifully together with ```numpy``` arrays. For example let's define an array

````python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--')
````

in ```matplotlib``` you can plot several arrays simple by giving the arguments sequentially to ```matplotlib```. For example let's suppose we want to plot two set of arrays

```t, t``` 

so a linear relationship and

```t, t**2```

so a quadratic relationship. This can be easily done in this way

````python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs')
````

## Plotting with categorical variables

Sometimes you want to do types of plots that are based on categorical variables. For example let's define the following arrays

````python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
````

````python
plt.bar(names, values)
````

or with scatter plots

````python
plt.scatter(names, values)
````

or with lines

````python
plt.plot(names, values)
````

## Putting several plots together

To put several plots together in an array, is very easy with ```matplotlib```. For example to make a $1\times 2$ array (one line of $2$ images) you can simply achieve it with the ```subplot()``` command.

The ```subplot()``` command specifies ```numrows, numcols, plot_number``` where ```plot_number``` ranges from 1 to ```numrows*numcols```. The commas in the subplot command are optional if ```numrows*numcols<10```. So ```subplot(211)``` is identical to ```subplot(2, 1, 1)```.

````python
plt.subplot(121)
plt.bar(names, values)

plt.subplot(122)
plt.plot(names, values)
````

to make the figure of a specific size, say ```9x3``` you can use the following syntax (note the dimensions are in inches).

````python
plt.figure(figsize=(9, 3))
plt.subplot(121)
plt.bar(names, values)

plt.subplot(122)
plt.plot(names, values)
````

You can create multiple figures by using multiple figure() calls with an increasing figure number. Of course, each figure can contain as many axes and subplots as you want.

````python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default
````

To setup line properties there are three main ways (see for a detailed explanation here https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py), but the one I think is the easiest is to use keywords tags. For example to set linewidth equal to 2.0, you would simply do ```plt.plot(x, y, linewidth=2.0)```. Here are the main tags you can use

```
linestyle or ls = [ '-' | '--' | '-.' | ':' | 'steps' | ...]
linewidth or lw	= float value in points
marker =	[ '+' | ',' | '.' | '1' | '2' | '3' | '4' ]
markersize or ms = float
```
